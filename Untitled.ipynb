```python
def draw(name, good_list, bad_list, start_date='20130201', end_date='20201109'):
    l, x, start_index = get_p_list_kr(name, start_date, end_date)

    #smoothed = double_smooth(l, 0.02, 0.03)
    p = l[start_index:]
    ma_240 = ma(l, start_index)
    ma_60 = ma(l, start_index, k=60)
    ma_20 = ma(l, start_index, k=10)
    ma_20 = ma(ma(ma_20, 0, k=10), 0, k=10)
    x = x[start_index:]
    #print(ma_240)   
    #dr = double_smooth(derivative(ma_240), 0.03, 0.03)
    dr = ma(derivative(ma_240),0, k=60)
    dr = derivative_abs(dr)
    plt.figure(figsize=(10,5))
    plt.plot(x, ma_240)
    plt.plot(x, p)
    plt.plot(x, ma_20)
    for bad in bad_list:
        plt.axvline(x=bad, color='r', linestyle=':', linewidth=1)
    for good in good_list:
        plt.axvline(x=good, color='b', linestyle=':', linewidth=1)
    plt.yscale('log')
    plt.show()
    #plt.figure(figsize=(10,10))
    #plt.plot(x, dr)
    #plt.plot(x, np.zeros(len(dr)))
    #plt.yscale('log')
    #plt.show()
```

```python
s_list = ['119860', '214150', '230360', '067900', '102710', '051160', '139670', '033290']
start_date='20171010'
end_date='20201109'
for name in s_list:
    #try:
    good_list, bad_list = function.backtest(name, date(int(start_date[:4]),int(start_date[4:6]),int(start_date[6:8])), 
                                            date(int(end_date[:4]),int(end_date[4:6]),int(end_date[6:8])), country='kr')
    draw(name, good_list, bad_list, start_date, end_date)
    #except:
    #    print('Error')
```
