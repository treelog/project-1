```python
def criteria(p, ma_low, kal):
    discripency_rate_cr = (p[-1]-ma_low[-1])/ma_low[-1]
    dr = ma(derivative(ma_low),0, k=60)
    annualized_growth_rate = (1+dr[-1])**240-1
    dr = derivative_abs(dr)
    #print(discripency_rate_cr)
    #print(annualized_growth_rate)
    if (discripency_rate_cr <annualized_growth_rate*1/4 and annualized_growth_rate > 0) :
        return True
    else:
        return False
    
def criteria4(p, ma_low, kal):
    discripency_rate_cr = (p[-1]-kal[-1])/kal[-1]
    dr = ma(derivative(ma_low),0, k=60)
    annualized_growth_rate = (1+dr[-1])**240-1
    if (discripency_rate_cr<0.05 and annualized_growth_rate > 0) :
        return True
    else:
        return False
    
def criteria5(p, ma_low, kal):
    x = p[-1] - ma_low[-1]
    kal = ma(kal, 0, k=10)
    dr = ma(derivative(kal),0)
    if x>0 and dr[-1]>0:
        return True
    else:
        return False
```

```python
stock_list = list(pd.read_csv('stock_list_kr.csv')['stock_name'])
stock_list = [s.split(':')[1] for s in stock_list]
start_date='20191010'
end_date='20201118'
k = 480
#stock_list = ['101360']
for name in stock_list:
    try:
        l, x, start_index = get_p_list_kr(name, start_date, end_date)
        p = l[start_index:]
        kalman = Kalman(Q=0.0002)
        kal = kalman(p)
        ma_low = ma(l, start_index, k=k)
        ma_low = ma(ma(ma_low, k=10), k=5)
        x = x[start_index:]
        for i in range(1, 2):
            if criteria4(p, ma_low, kal):
                x = ma(derivative(ma_low),0, k=60)[-i]
                annualized_growth_rate = (1+x)**240-1
                print(stock.get_market_ticker_name(name), annualized_growth_rate)
                break
    except:
        print(name, 'Error')
```
